1. Load Your Excel Data

```python
import pandas as pd

df = pd.read_excel('dataset_manual_statistics.xlsx')
```

```python
features = ['# Words', '# Sentences', '# Pools', '# Lanes', '# Data Objects',
            '# Decision Points', '# Parallel', '# Loops',  'Depth']
```

2. Preprocess the Data

```python
# Function to clean and sum the lane numbers
def extract_lane_count(val):
    if isinstance(val, str):
        nums = [int(s.strip()) for s in val.replace('->', '').split('\n') if s.strip().isdigit()]
        return sum(nums)
    return val

# Apply the function
df['# Lanes'] = df['# Lanes'].apply(extract_lane_count)
```

```python
import re

def clean_numeric(val):
    # Convert non-string values to string for safe regex
    val_str = str(val)

    # Remove anything that's not a digit or dot
    val_str = re.sub(r'[^\d.]', '', val_str)

    # If it ends up empty or invalid, return 0
    if val_str == '' or val_str == '.':
        return 0

    try:
        return float(val_str)
    except ValueError:
        return 0



for col in features:
    df[col] = df[col].apply(clean_numeric)
```

```python
X = df[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

3. Cluster the Processes

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

4. Save clustring result

```python
df.to_excel('clustering_results.xlsx', index=False)
```
